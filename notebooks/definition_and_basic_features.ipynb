# Explorando os dados para definir os parâmetros de predição

Leitura dos arquivos .csv de pacientes, condições e encontros. As colunas referentes a datas também foram convertidas para *datetime*.

```python
import pandas as pd
import numpy as np

#Cenário 01
path01 = 'https://github.com/caiops/P2_IA368X/raw/main/data/external/scenario01/'

patients01 = pd.read_csv(path01 + 'patients.csv')
patients01['BIRTHDATE'] = pd.to_datetime(patients01['BIRTHDATE'])
patients01['DEATHDATE'] = pd.to_datetime(patients01['DEATHDATE'])

conditions01 = pd.read_csv(path01 + 'conditions.csv')
conditions01['START'] = pd.to_datetime(conditions01['START'])
conditions01['STOP'] = pd.to_datetime(conditions01['STOP'])

encounters01 = pd.read_csv(path01 + 'encounters.csv')
encounters01['START'] = pd.to_datetime(encounters01['START'])
encounters01['STOP'] = pd.to_datetime(encounters01['STOP'])

#Cenário 02
path02 = 'https://github.com/caiops/P2_IA368X/raw/main/data/external/scenario02/'

patients02 = pd.read_csv(path02 + 'patients.csv')
patients02['BIRTHDATE'] = pd.to_datetime(patients02['BIRTHDATE'])
patients02['DEATHDATE'] = pd.to_datetime(patients02['DEATHDATE'])

conditions02 = pd.read_csv(path02 + 'conditions.csv')
conditions02['START'] = pd.to_datetime(conditions02['START'])
conditions02['STOP'] = pd.to_datetime(conditions02['STOP'])

encounters02 = pd.read_csv(path02 + 'encounters.csv')
encounters02['START'] = pd.to_datetime(encounters02['START'])
encounters02['STOP'] = pd.to_datetime(encounters02['STOP'])
```

Primeiro, verificamos os tipos e quantidade de encontros únicos disponíveis com base em sua descrição. Em especial, verificamos os encontros do tipo *emergency*.

```python
#Descrições únicas de encontros no cenário 01
encounters01_desc = encounters01['DESCRIPTION'].value_counts()
for d in encounters01_desc.index:
  print(d, encounters01_desc[d])
```

```python
#Descrições únicas de encontros no cenário 02
encounters02_desc = encounters02['DESCRIPTION'].value_counts()
for d in encounters02_desc.index:
  print(d, encounters02_desc[d])
```

```python
#Descrições únicas de encontros do tipo "emergency" no cenário 01
encounters01_emerg_desc = encounters01.loc[encounters01['ENCOUNTERCLASS'] == 'emergency', 'DESCRIPTION'].value_counts()
for d in encounters01_emerg_desc.index:
  print(d, encounters01_emerg_desc[d])
```

```python
#Descrições únicas de encontros do tipo "emergency" no cenário 02
encounters02_emerg_desc = encounters02.loc[encounters02['ENCOUNTERCLASS'] == 'emergency', 'DESCRIPTION'].value_counts()
for d in encounters02_emerg_desc.index:
  print(d, encounters02_emerg_desc[d])
```

Então, verificamos os tipos e quantidade de condições únicas disponíveis com base em sua descrição.

```python
#Descrições únicas de condições no cenário 01
conditions01_desc = conditions01['DESCRIPTION'].value_counts()
for d in conditions01_desc.index:
  print(d, conditions01_desc[d])
```

```python
#Descrições únicas de condições no cenário 02
conditions02_desc = conditions02['DESCRIPTION'].value_counts()
for d in conditions02_desc.index:
  print(d, conditions02_desc[d])
```

A neutropenia febril (NF) - descrita no conjunto de dados como *Febrile neutropenia (disorder)*, código 409089005 - chamou nossa atenção e decidimos investigar um pouco mais.

```python
#Pegando os dados de condição para os pacientes com NF
nf01 = conditions01.loc[conditions01['CODE'] == 409089005]
display(nf01)

nf02 = conditions02.loc[conditions02['CODE'] == 409089005]
display(nf02)
```

```python
print('Contagem de pacientes únicos com NF (cenário 01):')
print(nf01['PATIENT'].value_counts())
print('\nNúmero de pacientes com NF (cenário 01):')
print(len(nf01['PATIENT'].value_counts()))

print('\nContagem de pacientes únicos com NF (cenário 02):')
print(nf02['PATIENT'].value_counts())
print('\nNúmero de pacientes com NF (cenário 02):')
print(len(nf02['PATIENT'].value_counts()))
```

Podemos ver que, nos dois cenários, cada paciente apresentou a condição apenas uma vez. Verificamos, então, os pacientes que morreram ou não.

```python
#Transformando o identificador dos pacientes em lista pra usar no query
nf01_pl = nf01['PATIENT'].value_counts().index
nf02_pl = nf02['PATIENT'].value_counts().index
#Pegando apenas os pacientes de interesse
patients01_i = patients01.query("Id in @nf01_pl")
patients02_i = patients02.query("Id in @nf02_pl")
#Visualizando
display(patients01_i)
display(patients02_i)
```

```python
#Dos pacientes de interesse, aqueles que morreram (se for NaT, ou seja, não morreu, a igualdade não vai ser respeitada)
patients01_i_d = patients01_i.loc[patients01_i['DEATHDATE'] == patients01_i['DEATHDATE']]
patients02_i_d = patients02_i.loc[patients02_i['DEATHDATE'] == patients02_i['DEATHDATE']]

print('Número de pacientes com NF que morreram (cenário 01):')
print(len(patients01_i_d))
print('\nNúmero de pacientes com NF que morreram (cenário 02):')
print(len(patients02_i_d))

#Visualizando
print()
display(patients01_i_d)
print()
display(patients02_i_d)
```

```python
#Transformando a coluna de 'Id' numa lista pro query
patients01_i_d_id = patients01_i_d['Id'].tolist()
patients02_i_d_id = patients02_i_d['Id'].tolist()
#Vendo os dados de NF pros que morreram
display(nf01.query('PATIENT in @patients01_i_d_id'))
print()
display(nf02.query('PATIENT in @patients02_i_d_id'))
```

```python
#Vendo os dados de NF pros que não morreram
display(nf01.query('PATIENT not in @patients01_i_d_id'))
display(nf02.query('PATIENT not in @patients02_i_d_id'))
```

Analisando as tabelas, tudo indica que os pacientes para os quais a NF não teve um fim (*STOP == NaT*) morreram no mesmo dia que começou a NF. Para aqueles que a NF passou e morreram, eles morreram só anos depois.

Verificamos, então, se isso realmente ocorre para todos os pacientes.

```python
print('Cenário 01\n')
#Pacientes que morreram com NF
nf01_d = nf01.loc[nf01['STOP'] != nf01['STOP']]
print('Número de pacientes que morreram com NF:')
print(len(nf01_d))
#Gerando uma lista com o identificador desses pacientes
nf01_d_pl = nf01_d['PATIENT'].tolist()

#Contando quantos desses pacientes realmente morreram no dia em que a NF começou e imprimindo aqueles em que isso não ocorreu
cont = 0
print('\nPacientes que não morreram no dia em que a NF começou:')
for p in nf01_d_pl:
  if nf01_d[nf01_d['PATIENT'] == p]['START'].tolist() == patients01_i[patients01_i['Id'] == p]['DEATHDATE'].tolist():
    cont += 1
  else:  
    print(p)

print('\nNúmero de pacientes que morreram no dia em que a NF começou:')
print(cont)
```

```python
#Verificando os dados do paciente que não morreu no dia em que a NF começou:
display(nf01_d[nf01_d['PATIENT'] == 'eabfa666-29b5-64f3-105a-e8ea7653ae70'])
display(patients01_i[patients01_i['Id'] == 'eabfa666-29b5-64f3-105a-e8ea7653ae70'])
```

```python
print('Cenário 02\n')
#Pacientes que morreram com NF
nf02_d = nf02.loc[nf02['STOP'] != nf02['STOP']]
print('Número de pacientes que morreram com NF:')
print(len(nf02_d))
#Gerando uma lista com o identificador desses pacientes
nf02_d_pl = nf02_d['PATIENT'].tolist()

#Contando quantos desses pacientes realmente morreram no dia em que a NF começou e imprimindo aqueles em que isso não ocorreu
cont = 0
print('\nPacientes que não morreram no dia em que a NF começou:')
for p in nf02_d_pl:
  if nf02_d[nf02_d['PATIENT'] == p]['START'].tolist() == patients02_i[patients02_i['Id'] == p]['DEATHDATE'].tolist():
    cont += 1
  else:  
    print(p)

print('\nNúmero de pacientes que morreram no dia em que a NF começou:')
print(cont)
```

```python
print('Cenário 01\n')
#Verificando quanto tempo depois da NF os outros pacientes morreram
#Todos que tiveram NF e morreram em algum momento
nf01_d_a = nf01.query('PATIENT in @patients01_i_d_id')
#Filtrando só aqueles que não morreram no momento da NF
nf01_d_a = nf01_d_a.loc[nf01_d_a['STOP'] == nf01_d_a['STOP']]
nf01_d_a_pl = nf01_d_a['PATIENT'].tolist()

#Verificando a diferença de dias entre o término da NF e a morte
print('Quantos dias depois da NF os outros pacientes morreram?')
for p in nf01_d_a_pl:
  print((patients01_i[patients01_i['Id'] == p]['DEATHDATE'].tolist()[0] - nf01_d_a[nf01_d_a['PATIENT'] == p]['STOP'].tolist()[0]).days)
```

```python
print('Cenário 02\n')
#Verificando quanto tempo depois da NF os outros pacientes morreram
#Todos que tiveram NF e morreram em algum momento
nf02_d_a = nf02.query('PATIENT in @patients02_i_d_id')
#Filtrando só aqueles que não morreram no momento da NF
nf02_d_a = nf02_d_a.loc[nf02_d_a['STOP'] == nf02_d_a['STOP']]
nf02_d_a_pl = nf02_d_a['PATIENT'].tolist()

#Verificando a diferença de dias entre o término da NF e a morte
print('Quantos dias depois da NF os outros pacientes morreram?')
for p in nf02_d_a_pl:
  print((patients02_i[patients02_i['Id'] == p]['DEATHDATE'].tolist()[0] - nf02_d_a[nf02_d_a['PATIENT'] == p]['STOP'].tolist()[0]).days)
```

Confirmamos que, de fato, os pacientes que morreram com NF o fizeram no dia em que a NF começou. A única exceção foi um paciente no cenário 01 que morreu um dia após o início da NF. Ainda, para aqueles que a NF passou e morreram, o intervalo entre o fim da NF e a morte do paciente foi de no mínimo 202 dias.

# Primeiros dados básicos para predição

Começamos, então, a criar as tabelas para usar no orange. Pegamos os dados dos pacientes que consideramos interessantes para realizar a classificação. Também renomeamos a coluna "Id" para "PATIENT" para facilitar as operações futuras com as outras tabelas. Depois, adicionamos as datas de início e fim da NF.

```python
#Dados de interesse dos pacientes
scenario01 = patients01_i[['Id', 'BIRTHDATE', 'RACE', 'ETHNICITY', 'GENDER']]
scenario01 = scenario01.rename(columns={'Id': 'PATIENT'})

scenario02 = patients02_i[['Id', 'BIRTHDATE', 'RACE', 'ETHNICITY', 'GENDER']]
scenario02 = scenario02.rename(columns={'Id': 'PATIENT'})

#Adicionando as datas de início e fim da NF
scenario01 = scenario01.merge(nf01[['PATIENT', 'START', 'STOP']], on='PATIENT', how='inner')
display(scenario01)

scenario02 = scenario02.merge(nf02[['PATIENT', 'START', 'STOP']], on='PATIENT', how='inner')
display(scenario02)
```

A partir da data do fim da NF, definimos uma nova coluna que indica se o paciente morreu ou não com a NF - ou seja, o nosso *target*.

```python
scenario01['DEATH_FN'] = scenario01['STOP'].apply(lambda x: 0 if x == x else 1)
scenario01 = scenario01.drop(['STOP'], axis=1)
display(scenario01)

scenario02['DEATH_FN'] = scenario02['STOP'].apply(lambda x: 0 if x == x else 1)
scenario02 = scenario02.drop(['STOP'], axis=1)
display(scenario02)
```

A partir da data de início da NF e da data de nascimento dos pacientes, definimos a idade na qual eles apresentaram a NF, inicialmente em dias.

```python
scenario01['AGE_FN_DAYS'] = (scenario01['START'] - scenario01['BIRTHDATE']).dt.days
scenario01 = scenario01.drop(['START'], axis=1)
display(scenario01)

scenario02['AGE_FN_DAYS'] = (scenario02['START'] - scenario02['BIRTHDATE']).dt.days
scenario02 = scenario02.drop(['START'], axis=1)
display(scenario02)
```

Verificamos, então, se existe alguma diferença significativa entre a idade em que o paciente recebeu o diagnóstico de câncer e a idade em que ele apresentou a NF.

A partir da data em que o diagnóstico de câncer começou e da data de nascimento do paciente, definimos sua idade no momento desse diagnóstico, também em dias.

```python
#Pegando as datas de início do diagnóstico de câncer
scenario01_t = scenario01.merge(amld01[['PATIENT', 'START']], on='PATIENT', how='inner')
display(scenario01_t)

scenario02_t = scenario02.merge(amld02[['PATIENT', 'START']], on='PATIENT', how='inner')
display(scenario02_t)
```

```python
#Obtendo as idades no momento do diagnóstico
scenario01_t['AGE_AMLD_DAYS'] = (scenario01_t['START'] - scenario01_t['BIRTHDATE']).dt.days
scenario01_t = scenario01_t.drop(['START'], axis=1)
display(scenario01_t)

scenario02_t['AGE_AMLD_DAYS'] = (scenario02_t['START'] - scenario02_t['BIRTHDATE']).dt.days
scenario02_t = scenario02_t.drop(['START'], axis=1)
display(scenario02_t)
```

```python
#Verificando os pacientes com diferença de idade entre os dois diagnósticos (câncer e NF)
for index, row in scenario01_t.iterrows():
    if row['AGE_AMLD_DAYS'] != row['AGE_FN_DAYS']:
      display(scenario01_t.loc[index])

print()

for index, row in scenario02_t.iterrows():
    if row['AGE_AMLD_DAYS'] != row['AGE_FN_DAYS']:
      display(scenario02_t.loc[index])
```

Podemos ver que poucos pacientes apresentaram diferença entre as duas idades. Além disso, os que apresentaram a diferença, ela foi de apenas 1 dia. Portanto, desconsideramos a data do diagnóstico do câncer para o nosso modelo.

Convertemos a idade em que os pacientes tiveram a NF de dias para anos e verificamos as idades mínima e máxima em cada cenário.

```python
scenario01['AGE_FN_YEARS'] = scenario01['AGE_FN_DAYS']//365
scenario01 = scenario01.drop(['AGE_FN_DAYS'], axis=1)
display(scenario01)

scenario02['AGE_FN_YEARS'] = scenario02['AGE_FN_DAYS']//365
scenario02 = scenario02.drop(['AGE_FN_DAYS'], axis=1)
display(scenario02)
```

```python
print('Idades máxima e mínima (cenário 01):')
print(scenario01['AGE_FN_YEARS'].min())
print(scenario01['AGE_FN_YEARS'].max())

print('\nIdades máxima e mínima (cenário 02):')
print(scenario02['AGE_FN_YEARS'].min())
print(scenario02['AGE_FN_YEARS'].max())
```
