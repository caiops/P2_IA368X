```python
def cond_30days(cond):
  sinusitis01 = conditions01_i_b.query("DESCRIPTION == @cond")
  sinusitis02 = conditions02_i_b.query("DESCRIPTION == @cond")
  sinusitis01_pl = sinusitis01['PATIENT'].value_counts().index
  sinusitis02_pl = sinusitis02['PATIENT'].value_counts().index

  print(len(sinusitis01_pl))
  print(len(sinusitis02_pl))

  #Pegando só os que tiveram no intervalo de um 30 dias antes da NF
  for p in sinusitis01_pl:
    date = nf01.loc[nf01['PATIENT'] == p, 'START'].values[0]
    date_i = date - np.timedelta64(30,'D')
    if not sinusitis01.query('PATIENT == @p & START >= @date_i & START <= @date').empty:
      display(sinusitis01.query('PATIENT == @p & START >= @date_i & START <= @date'))

  print()

  for p in sinusitis02_pl:
    date = nf02.loc[nf02['PATIENT'] == p, 'START'].values[0]
    date_i = date - np.timedelta64(30,'D')
    if not sinusitis02.query('PATIENT == @p & START >= @date_i & START <= @date').empty:
      display(sinusitis02.query('PATIENT == @p & START >= @date_i & START <= @date'))
```
