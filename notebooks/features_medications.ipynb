# Explorando a tabela de medicamentos para obter outras *features* interessantes

Leitura dos arquivos .csv de medicamentos, além dos arquivos com as *features* e dos dados de NF definidos através do notebook "features_conditions_encounters.ipynb". As colunas referentes a datas também foram convertidas para *datetime*.

```python
import pandas as pd
import numpy as np

#Cenário 01

medications01 = pd.read_csv('https://github.com/caiops/P2_IA368X/raw/main/data/external/scenario01/medications.csv')
medications01['START'] = pd.to_datetime(medications01['START'])
medications01['STOP'] = pd.to_datetime(medications01['STOP'])

scenario01 = pd.read_csv('https://github.com/caiops/P2_IA368X/raw/main/data/interim/features/features_conditions_encounters_01.csv')
scenario01['BIRTHDATE'] = pd.to_datetime(scenario01['BIRTHDATE'])

nf01 = pd.read_csv('https://github.com/caiops/P2_IA368X/raw/main/data/interim/interest/nf01.csv')
nf01['START'] = pd.to_datetime(nf01['START'])
nf01['STOP'] = pd.to_datetime(nf01['STOP'])

#Cenário 02

medications02 = pd.read_csv('https://github.com/caiops/P2_IA368X/raw/main/data/external/scenario02/medications.csv')
medications02['START'] = pd.to_datetime(medications02['START'])
medications02['STOP'] = pd.to_datetime(medications02['STOP'])

scenario02 = pd.read_csv('https://github.com/caiops/P2_IA368X/raw/main/data/interim/features/features_conditions_encounters_02.csv')
scenario02['BIRTHDATE'] = pd.to_datetime(scenario02['BIRTHDATE'])

nf02 = pd.read_csv('https://github.com/caiops/P2_IA368X/raw/main/data/interim/interest/nf02.csv')
nf02['START'] = pd.to_datetime(nf02['START'])
nf02['STOP'] = pd.to_datetime(nf02['STOP'])
```

Obtendo os dados de medicamento apenas para os pacientes de interesse.

```python
#Obtendo as listas de pacientes
patients01 = scenario01['PATIENT'].tolist()
patients02 = scenario02['PATIENT'].tolist()
#Obtendo as listas de pacientes de interesse que sobreviveram ou não à NF
death01 = scenario01[scenario01['DEATH_FN'] == 1]['PATIENT'].tolist()
death02 = scenario02[scenario02['DEATH_FN'] == 1]['PATIENT'].tolist()
survive01 = scenario01[scenario01['DEATH_FN'] == 0]['PATIENT'].tolist()
survive02 = scenario02[scenario02['DEATH_FN'] == 0]['PATIENT'].tolist()
```

```python
#Obtendo os dados para os pacientes de interesse
#Cenário 01
medications01_i = medications01.query('PATIENT in @patients01').copy()
#Cenário 02
medications02_i = medications02.query('PATIENT in @patients02').copy()
```

```python
#Visualizando os dados de medicamentos
#Cenário 01
display(medications01_i)
#Cenário 02
display(medications02_i)
```

Encontrando os dados de medicamento cuja data de início é anterior ou igual à data de início da NF.

```python
#Mudando as datas iniciais de prescrição do formato data/hora para apenas data, de modo a realizar a comparação seguinte
medications01_i['START'] = medications01_i['START'].dt.date
medications01_i['START'] = pd.to_datetime(medications01_i['START'])

medications02_i['START'] = medications02_i['START'].dt.date
medications02_i['START'] = pd.to_datetime(medications02_i['START'])
```

```python
#Obtendo os medicamentos prescritos aos pacientes de interesse antes ou no dia que a NF começou
medications01_i_b = pd.DataFrame(columns=['PATIENT', 'START', 'CODE', 'DESCRIPTION', 'DISPENSES', 'REASONCODE', 'REASONDESCRIPTION'])
for p in patients01:
  date = nf01.loc[nf01['PATIENT'] == p, 'START'].values[0]
  medications01_i_b = pd.concat([medications01_i_b, medications01_i.query('PATIENT == @p & START <= @date')[['PATIENT', 'START', 'CODE', 'DESCRIPTION', 'DISPENSES', 'REASONCODE', 'REASONDESCRIPTION']]], ignore_index=True)

display(medications01_i_b)

medications02_i_b = pd.DataFrame(columns=['PATIENT', 'START', 'CODE', 'DESCRIPTION', 'DISPENSES', 'REASONCODE', 'REASONDESCRIPTION'])
for p in patients02:
  date = nf02.loc[nf02['PATIENT'] == p, 'START'].values[0]
  medications02_i_b = pd.concat([medications02_i_b, medications02_i.query('PATIENT == @p & START <= @date')[['PATIENT', 'START', 'CODE', 'DESCRIPTION', 'DISPENSES', 'REASONCODE', 'REASONDESCRIPTION']]], ignore_index=True)

display(medications02_i_b)
```

Verificando a contagem de diferentes descrições.

```python
#Cenário 01
print(medications01_i_b['DESCRIPTION'].value_counts())
#Cenário 02
print()
print(medications02_i_b['DESCRIPTION'].value_counts())
```

Verificando o número total de medicamentos prescritos para cada paciente.

```python
#Cenário 01
med01 = medications01_i_b['PATIENT'].value_counts().reset_index()
med01.columns = ['PATIENT', 'MED_CONT']
display(med01)
#Cenário 02
med02 = medications02_i_b['PATIENT'].value_counts().reset_index()
med02.columns = ['PATIENT', 'MED_CONT']
print()
display(med02)
```

```python
#Cenário 01
med01 = med01.merge(scenario01[['PATIENT', 'DEATH_FN', 'AGE_FN_YEARS']], on='PATIENT', how='left')
display(med01)

#Cenário 02
med02 = med02.merge(scenario02[['PATIENT', 'DEATH_FN', 'AGE_FN_YEARS']], on='PATIENT', how='left')
display(med02)
```

```python
#Verificando de forma simples se existe uma diferença no número de mortos entre os que tomaram mais ou menos medicamentos
threshold = 3
print('Cenário 01\n')
med01_m = med01.loc[med01['MED_CONT'] >= threshold]
print(f'Número de pacientes que tomaram pelo menos {threshold} medicamentos:')
print(len(med01_m))
print(f'Número de mortos que tomaram pelo menos {threshold} medicamentos:')
print(med01_m['DEATH_FN'].sum())
med01_l = med01.loc[med01['MED_CONT'] < threshold]
print(f'\nNúmero de pacientes que tomaram menos de {threshold} medicamentos:')
print(len(med01_l))
print(f'Número de mortos que tomaram menos de {threshold} medicamentos:')
print(med01_l['DEATH_FN'].sum())

print('\nCenário 02\n')
med02_m = med02.loc[med02['MED_CONT'] >= threshold]
print(f'Número de pacientes que tomaram pelo menos {threshold} medicamentos:')
print(len(med02_m))
print(f'Número de mortos que tomaram pelo menos {threshold} medicamentos:')
print(med02_m['DEATH_FN'].sum())
med02_l = med02.loc[med02['MED_CONT'] < threshold]
print(f'\nNúmero de pacientes que tomaram menos de {threshold} medicamentos:')
print(len(med02_l))
print(f'Número de mortos que tomaram menos de {threshold} medicamentos:')
print(med02_l['DEATH_FN'].sum())
```

```python
#Verificando o número de pacientes para os quais foram prescritos remédios por conta de "Childhood asthma"
medications01_i_b.query("REASONDESCRIPTION == 'Childhood asthma'")[['PATIENT']].value_counts()
```

```python
#Verificando o número de pacientes para os quais foram prescritos remédios por conta de "Acute myeloid leukemia disease (disorder)" (código 91861009)
med01_nf = medications01_i_b.loc[medications01_i_b["REASONCODE"]== 91861009]
print("N° de pacintes com medicamento prescrito para Acute myeloid leukemia disease (disorder), cenário 01: ", len(med01_nf['PATIENT'].value_counts()))
print(med01_nf['PATIENT'].value_counts())
print('\nConfirmando o tipo de medicamento prescrito:')
print(med01_nf["DESCRIPTION"].value_counts())

med02_nf = medications02_i_b.loc[medications02_i_b["REASONCODE"]== 91861009]
print("\nN° de pacintes com medicamento prescrito para Acute myeloid leukemia disease (disorder), cenário 02: ", len(med02_nf['PATIENT'].value_counts()))
print(med02_nf['PATIENT'].value_counts())
print('\nConfirmando o tipo de medicamento prescrito:')
print(med02_nf["DESCRIPTION"].value_counts())
```

```python
#Verificando se todos os pacientes começaram a tomar o medicamento no mesmo dia em que receberam o diagnóstico de NF

print('Cenário 01\n')
#Gerando uma lista com o identificador dos pacientes
med01_nf_pl = med01_nf['PATIENT'].tolist()

#Contando quantos desses pacientes começaram a tomar o medicamento no dia em que a NF começou e imprimindo aqueles em que isso não ocorreu
cont = 0
print('Pacientes que não começaram a tomar o medicamento no dia em que a NF começou:')
for p in med01_nf_pl:
  if med01_nf[med01_nf['PATIENT'] == p]['START'].tolist() == nf01[nf01['PATIENT'] == p]['START'].tolist():
    cont += 1
  else:  
    print(p)

print('\nNúmero de pacientes que começaram a tomar o medicamento no dia em que a NF começou:')
print(cont)

print('\nCenário 02\n')
#Gerando uma lista com o identificador dos pacientes
med02_nf_pl = med02_nf['PATIENT'].tolist()

#Contando quantos desses pacientes começaram a tomar o medicamento no dia em que a NF começou e imprimindo aqueles em que isso não ocorreu
cont = 0
print('Pacientes que não começaram a tomar o medicamento no dia em que a NF começou:')
for p in med02_nf_pl:
  if med02_nf[med02_nf['PATIENT'] == p]['START'].tolist() == nf02[nf02['PATIENT'] == p]['START'].tolist():
    cont += 1
  else:  
    print(p)

print('\nNúmero de pacientes que começaram a tomar o medicamento no dia em que a NF começou:')
print(cont)
```

```python
#Acrescentando a informação se o paciente tomou ou não o medicamento para "Acute myeloid leukemia disease (disorder)" nas tabelas pro orange:
scenario01 = scenario01.merge(med01_nf[['PATIENT', 'DESCRIPTION']], on='PATIENT', how='left')

scenario01['MED_AMLD'] = scenario01['DESCRIPTION'].apply(lambda x: 0 if x != x else 1)
scenario01 = scenario01.drop(['DESCRIPTION'], axis=1)

display(scenario01)

scenario02 = scenario02.merge(med02_nf[['PATIENT', 'DESCRIPTION']], on='PATIENT', how='left')

scenario02['MED_AMLD'] = scenario02['DESCRIPTION'].apply(lambda x: 0 if x != x else 1)
scenario02 = scenario02.drop(['DESCRIPTION'], axis=1)

display(scenario02)
```
